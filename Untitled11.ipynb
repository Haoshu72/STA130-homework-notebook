Q5.
Hypothesis: The null hypothesis (H0) states that there is no inherent preference for head tilting direction when kissing. This hypothesis suggests that people are equally likely to tilt their heads to the left or the right, similar to a 50/50 chance in flipping a coin.



```python
import numpy as np
trials = 10000 
sample_size = 124  
observed_rights = 80  
simulated_data = np.random.binomial(n=sample_size, p=0.5, size=trials)
p_value = np.mean(simulated_data >= observed_rights)
print("P-value:", p_value)
```

Conclusion: Since the p value of 0.007 is significantly less than 0.05, we have enough evidence to reject the null hypothesis. This suggests that 80 out of 124 couples tend to tilt their heads to the right when kissing, which is unlikely to happen randomly. Therefore, we conclude that people are more likely to tilt their heads to the right than to the left when kissing.

A p-value is a probability between 0 and 1, which is a range rather than an exact value. It helps to evaluate the consistency of observational data with the null hypothesis. A smaller p-value indicates stronger evidence against the null hypothesis, while a larger P-value indicates weaker evidence. However, the P-value does not provide absolute certainty because it only measures the probability of the data under the assumption that the null hypothesis is true. There are also two possible errors: Type 1 errors (false positives), where we reject the null hypothesis when it is actually true; Type 2 error (false negative), when the null hypothesis is false we cannot reject it. In Fido's case, the P-value can only indicate the strength of the evidence, and cannot definitively prove innocence or non-innocence, since it is never absolute evidence.

Q8.Complete the following assignment.
(1) In this experiment, our null hypothesis H0 is that the proportion of students correctly identifying the order of milk and tea pouring is equal to 0.5, which indicates that their judgment is merely the result of random guessing.In contrast, the alternative hypothesis Ha is that the students were able to correctly identify the pour order at a ratio of less than 0.5, meaning that their judgment ability exceeded the level of random guessing.


```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
n_students = 80  # Total number of students
n_trials = 10000  # Number of simulations
correct_identifications = 49  # Number of correct identifications

# Perform simulations for the null hypothesis (p = 0.5)
simulated_correct_counts = np.random.binomial(n_students, 0.5, n_trials)

# Visualization of the simulation distribution
plt.figure(figsize=(10, 6))
plt.hist(simulated_correct_counts, bins=30, alpha=0.7, color='blue', edgecolor='black', density=True)
plt.axvline(correct_identifications, color='red', linestyle='dashed', linewidth=2, 
            label='Observed Correct Identifications (49)')
plt.title('Distribution of Simulated Correct Identifications under Null Hypothesis')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

P-value = 0.0442: Since this P-value is less than the common significance level (e.g. 0.05), we can reject the null hypothesis H0.This showed that the ability of the 49 observed students to correctly identify the pouring order of milk and tea was significantly higher than the level of random guessing. That is, students' ability to judge is not random, but through recognition.
